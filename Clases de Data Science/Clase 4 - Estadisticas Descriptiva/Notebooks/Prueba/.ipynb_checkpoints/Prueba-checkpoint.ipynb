```python
import numpy as np 
```

```python
filename="Formulario_C4.csv"
data_cruda = np.genfromtxt(filename,delimiter=",",dtype=str) # Notar que si no ponemos el separador correcto tira error. 
data_cruda

alturas = data_cruda[1:,1].astype(float) # Seleccionamos la columna 1 desde la fila uno hasta el final
print(alturas)

peso = data_cruda[1:,2].astype(float) # Seleccionamos la columna 1 desde la fila uno hasta el final
print(peso)
```

```python
# Calculamos la media, la mediana y la moda para las columnas
print(f'La estatura promedio del curso es:{np.mean(alturas)}')
print(f'El peso promedio del curso es:{np.mean(peso)}')

print(f'La mediana de la altura del curso es:{np.median(alturas)}')
print(f'La mediana del peso del curso es:{np.median(peso)}')
# numpy no tiene un metodo para calcular la moda!!!!!
```

```python
# Pero otra libreria como scipy si
from scipy import stats # No funciona import scipy y usar scipy.stats
print (f'La moda para las alturas del curso es: {stats.mode(alturas)}') # Vemos que no tiene mucho sentido!
```

```python
# Calcular rango intercuartil
print(f'El rango de altura va desde {np.max(alturas)} hasta {np.min(alturas)}.')
print(f'El rango de peso va desde {np.max(peso)} hasta {np.min(peso)}.')

# Ahora queremos saber donde esta el 50% de las alturas mas comunes
# Numpy tiene una funcion que separa en partes ordenando

print(np.percentile(alturas,[75,50,25]))

print (f'El 50% de las alturas mas comunes estan entre: {np.percentile(alturas, [75 ,25])[1]}cm y {np.percentile(alturas, [75 ,25])[0]}cm.')
```

```python
# Calcular varianza
print(f'La varianza en las alturas es de: {np.var(alturas)} centrimetros cuadrados')

# Calcular la desviacion estandar
print(f'La desviacion estandar es de:{np.std(alturas)} centimetros')
```

```python
# Vamos a calcular que porcentaje de los valores caen dentro de una desviacion estandar de la media

n = 2 # cambia de 1, 2 y 3
sd = np.std(alturas)
mean = np.mean(alturas)

mascara = (mean - n * sd < alturas) & (alturas < mean + n * sd)
print (alturas[mascara])
print (mascara.sum())
print (f'El porcenaje de valores que caen dentro del rango indicado es: {mascara.sum()/len(mascara)*100}%')
```

```python
def coef_var(data):
    return np.std(data)/np.mean(data)    
```

```python
# Comparando variabilidades entre dos variables

print (f'La desviacion estandar de las alturas es {np.std(alturas)}cm, y su valor medio es {np.mean(alturas)}cm. Su ceoficiente de variación es {coef_var(alturas)}')
print (f'La desviacion estandar de los peso es {np.std(peso)}Kg, y su valor medio es {np.mean(peso)}Kg. Su ceoficiente de variación es {coef_var(peso)}')
```
